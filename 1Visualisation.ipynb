```python
%matplotlib inline
import seaborn as sns
from matplotlib import cm
import pandas as pd
```

```python
fig = plt.figure(figsize=(27, 18))
axes = []
for idc, c in enumerate(df):
    axes.append(fig.add_subplot(2, 3, idc+1))
    axes[idc].set_title(c, fontsize=20)
    nonempty_df = df[c][df[c] != 0]
    nonzero_labels = df[c][df[c] != 0].keys()
    cleaned_labels = [l  if l in nonzero_labels else "" for l in df[c].keys()]
    patches = axes[idc].pie(df[c], colors=cm.Accent(np.arange(len(df[c]))/len(df[c])), labels=cleaned_labels)
#     print(set(df[c].keys())
```
